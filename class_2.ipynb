## Stacks

A Stack is an ordered collection of elements where items are added and removed from one end, commonly reffered to as the "top".  The ordering principle is reffered to as LIFO (Last in, First out)

```python
# Simplified version (relying on built-in functions)

class Stack:
    def __init__(self):
        self.items = []
        
    def push(self, elements):
        self.items.append(elements)
        
    def pop(self):
        return self.items.pop()
        
    def peek(self):
        return self.items[len(self.items)-1]
    
    def size(self):
        return len(self.items)
    
    def is_empty(self):
        return self.items == []
    
    def __str__(self):
        if self.items is None:
            print(self.items)
```

```python
stack_obj = Stack()

stack_obj.push("A")
stack_obj.push("B")
stack_obj.push("C")

top = stack_obj.pop()
print(stack_obj.items)
print(top)
print(stack_obj.peek())
print(stack_obj.size())
```

```python
# Custom Stack Class (from scratch)

class CustomStack():    # Creates a new class. A class is like a blueprint. Every time you create a stack from this class, Python creates a new stack object.
    
    class Node:         # Creates a class inside the CustomStack class. A Node is one individual item in your stack.
                        # Since this is a linked-list style stack, each item stores:
                        # 1) The actual data
                        # 2) A link to the item below it
        
        def __init__(self, data):       # Constructor for the Node class. It runs automatically whenever you create a new node.
                                        # self means the current node being created.
                                        # data means the value you want to store inside that node.
            self.data = data            # Stores the value that was passed into data.
            self.below = None           # Points to the node underneath the current node. Starts as None because nothing is below it yet.
    
    def __init__(self):                 # Constructor for the CustomStack class. It runs automatically when you create a stack.
        self.top = None                 # Keeps track of the top node in the stack. Starts as None because the stack is empty.
        
    def push(self, value):              # Creates the push() method. push means "add a new value to the top of the stack."
        if not self.top:                # Checks whether the stack is empty. If self.top is None, this is True.
            self.top = self.Node(value) # If the stack is empty, this creates a new node and makes it the top of the stack.
        else:                           # Runs if the stack is not empty.
            new_node = self.Node(value) # Creates a new node with the value you are pushing.
            new_node.below = self.top   # Links the new node to the current top node.
            self.top = new_node         # Updates the top pointer so the new node becomes the top of the stack.
            
    def pop(self):                      # Because pop removes the top item, it does not need a value parameter.
        if not self.top:                # Checks if the stack is empty.
            return "There is nothing to remove"
        else: 
            removed_value = self.top.data   # Saves the top node's data.
            self.top = self.top.below       # Moves top down one node.
            return removed_value            # Returns the removed value.
        
    def is_empty(self):
        return self.top is None         # Returns True if the stack has no top node.
        
# Assignment 2: Build the logic for the pop and is_empty functions

# def pop():

# def is_empty():
    
```
