# Class 3 - Queue

A queue is an ordered collection of elements where items are added and removed from **opposite** ends,
commonly referred to as the "front" and "back" or "rear" of the queue.  The ordering principle is referred 
to as FIFO (First in, First out) or "first come first served"

```python
# Simplified Queue (relying on built-ins)

class Queue:
    def __init__(self):
        self.items = []

def enqueue(self, element):
    self.items.insert(0,element)

def dequeue(self):
    return self.items.pop()

# Nice to have methods
def peek(self):
    return self.items[len(self.items)-1]

def size(self):
    return len(self.items)

def is_empty(self):
    return self.items == []
```
