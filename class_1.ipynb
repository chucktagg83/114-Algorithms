```python
# Expression - Operators and Operands
a = 10
b = 3

# Arithmetic Operators
print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division

# Relational Operators
print(a > b)  # Greater than
print(a < b)  # Less than
print(a >= b) # Greater than or equal to
print(a <= b) # Less than or equal to
print(a == b) # Equal to
print(a != b) # Not equal to

# Logical Operators (and, or, not)
c = True
d = False
print(c and d)  # Logical AND
print(c or d)   # Logical OR
print(not c)    # Logical NOT
```

```python
# Functions - A function is a reusable block of code that performs a specific task. In Python, you can define a function using the def keyword. For example:
def my_function():
    a = 10
    b = 3
    print(a + b)  # This will print the sum of a and b
    
def my_function_with_parameters(a,b):
    print(a + b) # This will print the sum of a and b
    
def my_function_with_return():
    a = 10
    b = 3
    return a + b  # This will return the sum of a and b

my_function()  # This will call the function and execute the code inside it
my_function_with_parameters(10, 3)  # This will call the function with parameters and execute the code inside it
result = my_function_with_return()  # This will call the function and store the returned value in the variable 'result'
print(result)  # This will print the value stored in the variable 'result'
```

```python
## Homework Assignement 1: Create a function that receives a collection and return an integer with the total numbers divisible by 2.  The collection is numbers = [3, 5, 27, 20, 6, 3.5, 60]

numbers = [3, 5, 27, 20, 6, 3.5, 60]  # A list of the numbers we are counting

def count_even_numbers(collection):   # Creates a function named count_even_numbers. I has one parameter called collection
    total = 0  # Starts the counter at 0
    
    for number in collection:  # Creates a loop through every number in collection
        if number % 2 == 0:  # checks if number is divisible by 2 with no remainder
            total += 1  # if number is divisible by 2, add 1 to the total
    
    return total   # Sends the final cunt back from the function

total = count_even_numbers(numbers)  # Runs the function using the numbers list

print("total = ", total)  # Prints the final answer
```

```python
## Homework Assignment 2: Implement the code for pseudocode challenge 2 with give list: nums = [12, 4, 27, 3, 1000, 50, 7]
nums = [12, 4, 27, 3, 1000, 50, 7]

nums = [12, 4, 27, 3, 1000, 50, 7]  # list of numbers

largest = nums[0]  # Stasrt by assuming the first number is the largest

for number in nums:  # loop through each number
    if number > largest:  # If current number is greater than largest
        largest = number # update largest to the current number

# Print the largest number
print(largest)  # print the largest number
```
